```python
import json
import matplotlib.pyplot as plt
import numpy as np
```

```python
methods = ["online_sac"]

fig, axes = plt.subplots(1, 2, figsize=(12, 6))

for i, method in enumerate(methods):
    with open(f'{method}_metrics.json', 'r') as f:
        metrics = json.load(f)
    evals = metrics['evaluations']

    ax = axes[i]
    ax.plot(np.arange(0, 5e6, 5000), evals, label=method)
    ax.set_xlabel('Number of Interactions')
    ax.set_ylabel('Average Return') if method == "sac" else ax.set_ylabel('Cumulative Future Rewards')
    ax.set_title('SAC') if method == "sac" else ax.set_title('Single-Episode SAC')
```
